#**Análise Dados do AirBnB - Rio de Janeiro**

##**O AirBnB**

O [AirBnB](www.airbnb.com.br), como se autoproclama em seu site, é uma comunidade de viagens. Atua no ramo de hospedagens, conectando pessoas que ofertam e demandam acomodações, apenas atuando a partir de sua plataforma online, mas sem possuir qualquer hotel. Seu objetivo é facilitar o encontro entre oferta e demanda de casas, quartos, hotéis e quaisquer outras propriedades do ramo hoteleiro.


O serviço nasceu em 2008, em San Francisco, Califórnia, de três amigos, Nathan Blecharczyk, Brian Chesky e Joe Gebbia, que ofertaram via web parte do apartamento onde moravam para hospedagem durante uma conferência na cidade enquanto a rede hoteleira local estava lotada. A partir dessa oportunidade, montaram um império hoje presente em mais de 191 países, hospedando milhões de pessoas por ano e fortalecendo a indústria do turismo cada vez mais. Recentemente, em dezembro de 2020, abriu seu capital na [NASDAQ](https://www.nasdaq.com/), captando 3,5 bilhões de dólares e atingindo o patamar de 100 bilhões de dólares de valor de mercado. Um belo salto em 12 anos.

Desde o início da pandemia do Covid-19 a plataforma tem voltado muito os seus esforços para manter as habitações cadastradas seguras para locação, tendo revisado sua política de segurança e padrões de higienização e de distanciamento entre hóspedes e anfitriões, tornando o negócio mais seguro e sustentável.

Entretanto, logo no início da análise foi identificada a presença de  *outliers*, o que fez necessário tratamento da base de dados para nova análise de distribuição com os mesmos tipos de gráfico. O desenvolvimento abaixo apresenta todo esse tratamento para chegar nas conclusões acima apresentadas (caso não queira acompanhar todo o desenvolvimento citado acima, pule para a questão 4).

```
# ver o resumo estatístico das variáveis numéricas
df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].describe()
```

```
# plotar o histograma das variáveis numéricas
df.hist(bins=15, figsize=(10,15));

```

**Obs**: É possível notar a presença de *outliers* em diversas variáveis, como em *price*, apresentando registros de até R$600 mil por dia, ou *minimum_nights* com até pelo menos 1000 noites para garantir a reserva.

```
# Plotar o boxplot das variáveis numéricas

#latitude
df.latitude.plot(kind='box', vert=False, figsize = (15,3))
plt.show()

#longitude
df.longitude.plot(kind='box', vert=False, figsize = (15,3))
plt.show()

#price
df.price.plot(kind='box', vert=False, figsize = (15,3))
plt.show()

#minimum_nights
df.minimum_nights.plot(kind='box', vert=False, figsize = (15,3))
plt.show()

#number_of_reviews
df.number_of_reviews.plot(kind='box', vert=False, figsize = (15,3))
plt.show()

#reviews_per_month
df.reviews_per_month.plot(kind='box', vert=False, figsize = (15,3))
plt.show()

#availability_365
df.availability_365.plot(kind='box', vert=False, figsize = (15,3))
plt.show()
```

**Obs**: o *boxplot* corrobora a presença de outliers, tratados abaixo.

**Preço**

Percebe-se, como já dito, a presença de *outliers* no preço porque a mais cara das hospedagens custa quase 600 mil reais a diária.

Como limite superior, **vamos estabelecer 5 mil reais a diária mais cara**, talvez uma casa grande a ser dividida para um grupo de muitas pessoas.

```
# ver a quantidade de valores acima de R$5000 para price
print("price: valores acima de R$5000:")
print("{} entradas".format(len(df[df.price > 5000])))
print("{:.4f}% ".format((len(df[df.price > 5000])/df.shape[0])*100))
```

Foram encontrados 438 casos acima de 5000 reais, ou 1,70% da base, a serem descartados.

```
# remover os preços considerados *outliers* em um novo DataFrame
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 5000].index, axis=0, inplace=True)
```

**Número mínimo de diárias** (*minimum_nights*)

Confome já discutido, também há *outliers* no número mínimo de diárias por haver registros de até 1000 dias nessa variável.
**Vamos estabelecer como limite do mínimo de diárias 1 mês**, o que pode ocorrer com um imóvel de temporada.

```
# ver a quantidade de valores acima de 30 dias para minimum_nights
print("minimum_nights: valores acima de 30:")
print("{} entradas".format(len(df[df.minimum_nights > 30])))
print("{:.4f}% ".format((len(df[df.minimum_nights > 30])/df.shape[0])*100))
```

Há 199 registros acima de 30 dias, ou 0,77% da base, também sendo descartados.

```
# remover mínimo de diárias considerados *outliers* em um novo DataFrame
df_clean.drop(df_clean[df_clean.minimum_nights > 30].index, axis=0, inplace=True)
```

**Demais variáveis**:

As demais variáveis ou não possuem *outliers* ou não serão utilizadas em nossa análise, não necessitando assim do mesmo tipo de tratamento.

**Análise sem outliers**

####**Q4. Há correlação entre as variáveis?**

Conforme o gráfico de calor abaixo, percebe-se que não há forte correlação entre as variáveis abaixo, com exceção das variáveis *number_of_reviews* e *reviews_per_month*, por se tratarem do mesmo tema.

```
# criar uma matriz de correlação
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()

# matriz de correlação *heatmap*

sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```
